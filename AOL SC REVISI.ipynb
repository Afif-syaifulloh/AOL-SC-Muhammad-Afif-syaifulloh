```python
#no 2 revisi
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

production_data = [
    1863, 1614, 2570, 1685, 2101, 1811, 2457, 2171, 2134, 2502, 2358, 2399, 
    2048, 2523, 2086, 2391, 2150, 2340, 3129, 2277, 2964, 2997, 2747, 2862, 
    3405, 2677, 2749, 2755, 2963, 3161, 3623, 2768, 3141, 3439, 3601, 3531, 
    3477, 3376, 4027, 3175, 3274, 3334, 3964, 3649, 3502, 3688, 3657, 4422, 
    4197, 4441, 4736, 4521, 4485, 4644, 5036, 4876, 4789, 4544, 4975, 5211, 
    4880, 4933, 5079, 5339, 5232, 5520, 5714, 5260, 6110, 5334, 5988, 6235, 
    6365, 6266, 6345, 6118, 6497, 6278, 6638, 6590, 6271, 7246, 6584, 6594, 
    7092, 7326, 7409, 7976, 7959, 8012, 8195, 8008, 8313, 7791, 8368, 8933, 
    8756, 8613, 8705, 9098, 8769, 9544, 9050, 9186, 10012, 9685, 9966, 10048, 
    10244, 10740, 10318, 10393, 10986, 10635, 10731, 11749, 11849, 12123, 
    12274, 11666, 11960, 12629, 12915, 13051, 13387, 13309, 13732, 13162, 
    13644, 13808, 14101, 13992, 15191, 15018, 14917, 15046, 15556, 15893, 
    16388, 16782, 16716, 17033, 16896, 17689
]

bulan = np.arange(1, 145)  # bikin array

# Fungsi untuk deret Taylor orde pertama (linear approximation)
def taylor_series_linear(x, x0, y0, dydx):
    return y0 + dydx * (x - x0)

# define bulan dan production 
x0 = bulan[0]
y0 = production_data[0]

dydx = (production_data[-1] - production_data[0]) / (bulan[-1] - bulan[0]) #turunan pertama

# Prediksi menggunakan deret Taylor linear
y_pred = taylor_series_linear(bulan, x0, y0, dydx)

plt.figure(figsize=(10, 6))
plt.plot(bulan, production_data, '.b', label='Data Produksi Bulanan')
plt.plot(bulan, y_pred, 'r', label='Taylor Series (Linear)')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Produksi')
plt.title('Prediksi Produksi Bulanan Tas dengan Deret Taylor (Linear Approximation)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# no 3
def p(x):
    return taylor_series_linear(x, x0, y0, dydx) - 25000

months_to_exceed_capacity = fsolve(p, 150)[0]  

start_build_month = months_to_exceed_capacity - 13

start_build_year = 2018 + (start_build_month // 12)
start_build_month_remainder = start_build_month % 12
start_build_month_name = ["Januari", "Februari", "Maret", "April", "Mei", "Juni", "Juli", "Agustus", "September", "Oktober", "November", "Desember"][int(start_build_month_remainder) - 1]

print(f"EGIER perlu memulai membangun gudang baru pada bulan {start_build_month_name} {start_build_year:.0f}.")
```
